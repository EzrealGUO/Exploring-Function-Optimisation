```python
import torch
import pandas as pd
df = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases'+'/iris/iris.data',header=None)
df = df.sample(frac = 1, random_state=0)

df = df[df[4].isin(['Iris-virginica','Iris-versicolor'])]#filter

#add lable indices column
mapping = {k:v for v,k in enumerate(df[4].unique())}
df[5] = (2*df[4].map(mapping))-1 #lablel in {-1,1}

print(df)

#normalisation
alldata = torch.tensor(df.iloc[:,[0,1,2,3]].values, dtype=torch.float)
alldata = (alldata - alldata.mean(dim=0))/alldata.var(dim=0)

#create datasets
targets_train = torch.tensor(df.iloc[:75,5].values, dtype=torch.long)
targets_valid = torch.tensor(df.iloc[75:,5].values, dtype=torch.long)
data_train = alldata[:75]
data_valid = alldata[75:]
print(data_train.size())
print(data_valid.size())
```

```python
def svm(x, w, b):
    h = (w*x).sum(1) + b
    return h
def hinge_loss(y_pred, y_true):
    N = y_true.size()[0]
    l = 1 - y_pred*y_true
    O = torch.zeros(N)
    loss = torch.max(l,O)
    e = torch.sum(loss)/N
    return e
```

```python
def hinge_loss(y_pred, y_true):
    hinge_loss = torch.ones_like(y_pred) - torch.mul(y_pred, y_true)
    hinge_loss = torch.max(torch.zeros_like(hinge_loss), hinge_loss)
    return torch.sum(hinge_loss)
```

```python
from torch.utils import data

dataset = data.TensorDataset(data_train,targets_train) # create your datset
dataloader = data.DataLoader(dataset, batch_size=25, shuffle=True) # create your dataloader
```

```python
w_SGD = torch.randn(1, 4, requires_grad=True)
b_SGD = torch.randn(1, requires_grad=True)

print(w_SGD)
print(b_SGD)
opt = optim.SGD([w_SGD,b_SGD], lr=0.01, weight_decay=0.0001)
SVM_SGD = np.random.randn(100,2)

for epoch in range(100):
    for batch in dataloader:
        opt.zero_grad()
        H_SGD = svm(data_train, w_SGD, b_SGD)
        loss_SGD = hinge_loss(H_SGD, targets_train)
        loss_SGD.backward()
        opt.step()
    SVM_SGD[epoch][0] = epoch+1
    SVM_SGD[epoch][1] = loss_SGD.data
```

```python
w_Adam = torch.randn(1, 4, requires_grad=True)
b_Adam = torch.randn(1, requires_grad=True)

print(w_Adam)
print(b_Adam)
opt = optim.Adam([w_Adam,b_Adam], lr=0.01, weight_decay=0.0001)
SVM_Adam = np.random.randn(100,2)


for epoch in range(100):
    for batch in dataloader:
        opt.zero_grad()
        H_Adam = svm(data_train, w_Adam, b_Adam)
        loss_Adam = hinge_loss(H_Adam, targets_train)
        loss_Adam.backward()
        opt.step()
    SVM_Adam[epoch][0] = epoch+1
    SVM_Adam[epoch][1] = loss_Adam.data
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(SVM_SGD[:,0],SVM_SGD[:,1],color = "red",label='SVM_SGD',linewidth=2)
ax.plot(SVM_Adam[:,0],SVM_Adam[:,1],color = "blue",label='SVM_Adam',linewidth=2)
ax.legend()
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

ax.set_xlim((0, 100))
ax.set_ylim((-10, 30))
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
w_random = torch.randn(1, 4, requires_grad=True)
b_random = torch.randn(1, requires_grad=True)
print(data_valid)
print(targets_valid)
output1 = svm(data_valid, w_SGD, b_SGD).data
output2 = svm(data_valid, w_Adam, b_Adam).data
output3 = svm(data_valid, w_random, b_random).data
print(output1)
print(output2)
print(output3)

for i in range(output1.size()[0]):
    if output1[i]>0:
        output1[i] = 1
    else:
        output1[i] =-1
        
for i in range(output2.size()[0]):
    if output2[i]>0:
        output2[i] = 1
    else:
        output2[i] =-1

for i in range(output2.size()[0]):
    if output3[i]>0:
        output3[i] = 1
    else:
        output3[i] =-1
        
print(output1)
print(output2)
print(output3)

Accuracy_SGD = accuracy_score(output1,targets_valid)
Accuracy_Adam = accuracy_score(output2,targets_valid)
Accuracy_Random = accuracy_score(output3,targets_valid)
print("Accuracy_SGD:",Accuracy_SGD)
print("Accuracy_Adam:",Accuracy_Adam)
print("Accuracy_Random:",Accuracy_Random)
```
